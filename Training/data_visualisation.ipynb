```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dataset_path = './Datasets/FERplus'
types = ['train','test']
for type_ in types:
    path_ = os.path.join(dataset_path, type_)
    for emotion in os.listdir(path_):
        emotion_path = os.path.join(path_, emotion)
        emotion_count = len(os.listdir(emotion_path))
        barg = plt.bar(emotion,emotion_count)
        plt.bar_label(barg) 
        
    plt.title(f'{type_.capitalize()} Dataset FER+')
    plt.xticks(rotation = 20)
    plt.show()
```
